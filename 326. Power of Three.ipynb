```python
n = 9

if n <= 0:
    print(False)
while n % 3 == 0:
    n //= 3

if n == 1:
    print(True)
else:
    print(False)
```

### Using Recursion

```python
def powerofthree(n):

    if n <= 0:
        return False
    
    if n == 1:
        return True
    
    if n % 3 != 0:
        return False
    
    return powerofthree(n // 3)

powerofthree(30)
```
